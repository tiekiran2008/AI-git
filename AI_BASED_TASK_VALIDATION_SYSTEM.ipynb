```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
from docx import Document
from bs4 import BeautifulSoup
import nbformat

def extract_docx(file):
    doc = Document(file)
    text = ""
    for para in doc.paragraphs:
        text += para.text + "\n"
    return text

def extract_txt(file):
    return open(file, encoding='utf-8').read()

def extract_html(file):
    soup = BeautifulSoup(open(file, encoding='utf-8'), 'html.parser')
    return soup.get_text()

def extract_code(file):
    return open(file, encoding='utf-8').read()

def extract_ipynb(file):
    nb = nbformat.read(file, as_version=4)
    text = ""
    for cell in nb.cells:
        if cell.cell_type == "code":
            text += cell.source
    return text
```

```python
import os

def pre_validate(file):
    if not os.path.exists(file):
        return False, "File not found"
    if os.path.getsize(file) == 0:
        return False, "Empty file"
    return True, "OK"
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_embedding(text):
    return model.encode(text)

def get_similarity(text1, text2):
    emb1 = get_embedding(text1)
    emb2 = get_embedding(text2)
    return cosine_similarity([emb1], [emb2])[0][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_embedding(text):
    return model.encode(text)

def get_similarity(text1, text2):
    emb1 = get_embedding(text1)
    emb2 = get_embedding(text2)
    return cosine_similarity([emb1], [emb2])[0][0]
```

```python
def check_result(score):
    if score >= 0.75:
        return "Valid ✅"
    elif score >= 0.60:
        return "Partially Valid ⚠️"
    else:
        return "Invalid ❌"
```

```python
def run_project(task_file, submission_file):

    # Check file
    ok, msg = pre_validate(submission_file)
    if not ok:
        return msg

    # Read task file
    if task_file.endswith('.docx'):
        task_text = extract_docx(task_file)
    else:
        task_text = extract_txt(task_file)

    # Read submission file
    if submission_file.endswith('.txt'):
        sub_text = extract_txt(submission_file)
    elif submission_file.endswith('.html'):
        sub_text = extract_html(submission_file)
    elif submission_file.endswith('.css') or submission_file.endswith('.js'):
        sub_text = extract_code(submission_file)
    elif submission_file.endswith('.ipynb'):
        sub_text = extract_ipynb(submission_file)
    elif submission_file.endswith('.docx'):
        sub_text = extract_docx(submission_file)

    if len(sub_text) < 100:
        return "Text too short ❌"

    # AI similarity
    score = get_similarity(task_text, sub_text)

    result = check_result(score)

    return f"Similarity: {score:.2f} → {result}"
```

```python
print(run_project("AI_Based_Task_validation_guide.docx", "submission.txt"))
```
